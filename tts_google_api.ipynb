```python
from google.cloud import texttospeech
import os
import tempfile
```

```python
UPLOAD_FOLDER = "output"
```

```python
def speak_text_google_return_url(text,lc = 'th-TH',voice_name='th-TH-Chirp3-HD-Despina') -> str: #th-TH-Chirp3-HD-Fenrir
    client_tts = texttospeech.TextToSpeechClient()
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.LINEAR16,
        effects_profile_id=["small-bluetooth-speaker-class-device"],
        pitch= 0,
        speaking_rate= 1.0
    )
    print(f"TTS (len txt) : {len(text)}")
    if len(text) > 500 :
        text = text[:500]
    
    voice = texttospeech.VoiceSelectionParams(
        language_code=lc,
        name=voice_name,
        ssml_gender=texttospeech.SsmlVoiceGender.MALE)
    synthesis_input = texttospeech.SynthesisInput(text=text)
    response = client_tts.synthesize_speech(
        input=synthesis_input, voice=voice, audio_config=audio_config
    )

    # Save in static directory to serve via Flask
    folder_path = os.path.join(os.getcwd(), UPLOAD_FOLDER,"tts_sound")
    os.makedirs(folder_path, exist_ok=True)

    with tempfile.NamedTemporaryFile(delete=False, suffix=".wav", dir=folder_path) as fp:
        fp.write(response.audio_content)
        tmp_filename = os.path.basename(fp.name)  # Just the file name

    # Return relative URL for browser use
    return f"/audio/{tmp_filename}"
```

```python
speak_text_google_return_url("ขออภัยค่ะ ฮู้ยังไม่สามารถตอบคำถามนี้ได้ค่ะ")
```
